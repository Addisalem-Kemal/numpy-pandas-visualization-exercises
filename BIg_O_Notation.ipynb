```python
import math
import matplotlib.pyplot as plt
```

```python
# Title your chart "Big O Notation"
# Label your x axis "Elements"
# Label your y axis "Operations"
# Label your curves or make a legend for the curves
# Use LaTex notation where possible


x = range(1, 20)

# y = 0(n) + 1 and label the curve "O(1)"

y = [0*x + 1 for x in x]

# y = log(n) and label the curve "O(log n)"

y1 = [math.log(x) for x in x]

# y = n and label the curve "O(n)"

y2 = x

# y = n*log(n) and label it "O(n logn)"

y3 = [x * math.log(x) for x in x]

# y = n^2 and label it "O(n^2)"

y4= [x**2 for x in x]

# y = 2^n and label it "O(n^2)"

y5 = [2**x for x in x]

# y = n! and label it "O(n!)"

y6 = [math.factorial(x) for x in x]

# y = n^n and label it "O(n^2)"

y7 =  [x**x for x in x]
```

```python
plt.figure(figsize = (16, 8))

plt.title('Big O Notation', fontsize=20)
plt.xlabel('Elements', fontsize=14)
plt.ylabel('Operations', fontsize = 14)
plt.xlim(1, 11)
plt.ylim(0, 50)
plt.plot(x, y, label= '$\mathcal{O}(1)$', color= 'red')
plt.plot(x, y1, label='$\mathcal{O}(logn)$', color= 'yellow')
plt.plot(x, y2, label='$\mathcal{O}(n)$', color= 'green')
plt.plot(x, y3, label='$\mathcal{O}(nlogn)$', color= 'blue')
plt.plot(x, y4, label='$\mathcal{O}(n^{2})$', color= 'salmon')
plt.plot(x, y5, label='$\mathcal{O}(2^{n})$', color= 'Purple')
plt.plot(x, y6, label='$\mathcal{O}(n!)$', color= 'orange')
plt.plot(x, y7, label='$\mathcal{O}(n^{n})$', color= 'brown')
plt.legend()
plt.show()
```
