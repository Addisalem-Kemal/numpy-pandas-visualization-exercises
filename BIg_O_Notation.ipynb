```python
import math
import matplotlib.pyplot as plt
```

```python

x = range(1, 20)

y = [0*x + 1 for x in x]

y1 = [math.log(x) for x in x]

y2 = x

y3 = [x * math.log(x) for x in x]

y4= [x**2 for x in x]

y5 = [2**x for x in x]

y6 = [math.factorial(x) for x in x]

y7 =  [x**x for x in x]
```

```python
plt.figure(figsize = (16, 8))

plt.title('Big O Notation', fontsize=20)
plt.xlabel('Elements', fontsize=14)
plt.ylabel('Operations', fontsize = 14)
plt.xlim(1, 11)
plt.ylim(0, 50)
plt.plot(x, y, label= '$\mathcal{O}(1)$', color= 'red')
plt.plot(x, y1, label='$\mathcal{O}(logn)$')
plt.plot(x, y2, label='$\mathcal{O}(n)$')
plt.plot(x, y3, label='$\mathcal{O}(nlogn)$')
plt.plot(x, y4, label='$\mathcal{O}(n^{2})$')
plt.plot(x, y5, label='$\mathcal{O}(2^{n})$')
plt.plot(x, y6, label='$\mathcal{O}(n!)$')
plt.plot(x, y7, label='$\mathcal{O}(n^{n})$')
plt.legend()
plt.show()
```
